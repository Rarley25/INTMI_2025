```python
#13. Solicite uma matriz 3x3 preenchida pelo usuário. Verifique se ela é uma matriz identidade.
matrix_3x3 = []
print("Por favor, digite os 9 valores para a matriz 3x3 (linha por linha):")
for i in range(3):
    row = []
    for j in range(3):
        while True:
            try:
                value = int(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                row.append(value)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")
    matrix_3x3.append(row)

print("Matriz criada:")
for row in matrix_3x3:
    print(row)
```

```python
is_identity = True


if len(matrix_3x3) != 3 or any(len(row) != 3 for row in matrix_3x3):
    is_identity = False
else:
    for i in range(3):
        for j in range(3):
            if i == j:
                if matrix_3x3[i][j] != 1:
                    is_identity = False
                    break
            else:
                if matrix_3x3[i][j] != 0:
                    is_identity = False
                    break
        if not is_identity:
            break

if is_identity:
    print("A matriz é uma matriz identidade!")
else:
    print("A matriz NÃO é uma matriz identidade.")
```

```python
#14. Faça um programa que leia uma matriz 3x3 e verifique se ela é simétrica (igual à sua transposta).
matrix_3x3_sym = []
print("Por favor, digite os 9 valores para a matriz 3x3 (linha por linha):")
for i in range(3):
    row = []
    for j in range(3):
        while True:
            try:
                value = int(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                row.append(value)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")
    matrix_3x3_sym.append(row)

print("Matriz criada:")
for row in matrix_3x3_sym:
    print(row)
```

```python
is_symmetric = True


for i in range(len(matrix_3x3_sym)):
    for j in range(i + 1, len(matrix_3x3_sym[i])):
        if matrix_3x3_sym[i][j] != matrix_3x3_sym[j][i]:
            is_symmetric = False
            break
    if not is_symmetric:
        break

if is_symmetric:
    print("A matriz é simétrica!")
else:
    print("A matriz NÃO é simétrica.")
```

```python
#17. Dada uma matriz 3x3 com valores reais, calcule o determinante (sem usar bibliotecas externas).
matrix_3x3_det = []
print("Por favor, digite os 9 valores para a matriz 3x3 (linha por linha, podem ser reais):")
for i in range(3):
    row = []
    for j in range(3):
        while True:
            try:
                value = float(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                row.append(value)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número real.")
    matrix_3x3_det.append(row)

print("Matriz criada:")
for row in matrix_3x3_det:
    print(row)
```

```python
if len(matrix_3x3_det) != 3 or any(len(row) != 3 for row in matrix_3x3_det):
    print("A matriz não é 3x3. Não é possível calcular o determinante.")
else:

    a, b, c = matrix_3x3_det[0]
    d, e, f = matrix_3x3_det[1]
    g, h, i = matrix_3x3_det[2]

    determinant = a * (e * i - f * h) - b * (d * i - f * g) + c * (d * h - e * g)

    print(f"O determinante da matriz é: {determinant}")
```

```python
import math

def is_perfect_square(n):
    if n < 0:
        return False
    if n == 0:
        return True
    sqrt_n = int(math.sqrt(n))
    return sqrt_n * sqrt_n == n

perfect_squares_set = set()

for num in original_set:
    if is_perfect_square(num):
        perfect_squares_set.add(num)

print("Conjunto de quadrados perfeitos:", perfect_squares_set)
```

```python
max_variation = -1
product_with_max_variation = None

for product, prices in products_prices.items():
    if len(prices) > 1:
        variation = max(prices) - min(prices)
        if variation > max_variation:
            max_variation = variation
            product_with_max_variation = product

if product_with_max_variation:
    print(f"\nO produto com a maior variação de preço é '{product_with_max_variation}' com uma variação de R$ {max_variation:.2f}.")
else:
    print("\nNão foi possível encontrar um produto com variação de preço suficiente ou os dados estão incompletos.")
```

```python
#30. Implemente um programa que leia uma frase e armazene em um dicionário o número de ocorrências de cada palavra (ignorando letras maiúsculas/minúsculas).
sentence = input("Por favor, digite uma frase: ")
print(f"Frase digitada: '{sentence}'")
```

```python
import re
from collections import defaultdict

# Converter a frase para minúsculas e remover pontuação
cleaned_sentence = re.sub(r'[^\w\s]', '', sentence.lower())
words = cleaned_sentence.split()

word_counts = defaultdict(int)

for word in words:
    if word: # Garante que não estamos contando strings vazias após a limpeza
        word_counts[word] += 1

print("Número de ocorrências de cada palavra:", dict(word_counts))
```
